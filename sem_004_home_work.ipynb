2.1 Изучите распределение домов от наличия вида на набережную
Постройте график
Сделайте выводы

```python
data = df['waterfront'].value_counts()
data
```

```python
plt.pie(data.values, labels= ['нет вида', 'есть вид'], autopct='%1.1f%%')
plt.legend(data.index)
plt.title('Вид на набережную');
```

99% жилья не имеет вид на набережную.

```python
waterfront = df[df["waterfront"] == 1]["price"].mean().round(2)
not_waterfront = df[df["waterfront"] == 0]["price"].mean().round(2)
waterfront, not_waterfront
```

Средняя стоимость жилья с видом на набережную выше

2.2 Изучите распределение этажей домов

```python
data = df['floors'].value_counts()
sns.barplot(x = data.index, y = data)
plt.title('Распределение этажей домов')
plt.xlabel('Кол-во этажей')
plt.ylabel('Кол-во');

```


2.2 Изучите распределение состояния домов

```python
data = df['condition'].value_counts()
sns.barplot(x = data.index, y = data)
plt.title('Распределение состояние домов')
plt.xlabel('Состояние')
plt.ylabel('Кол-во');
```

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.

```python
data = df[['sqft_living', 'sqft_above', 'grade', 'bedrooms', 'bathrooms', 'price']]
data
```

```python
sns.pairplot(data,);
```

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg') 
sns.jointplot(x=df['price'], y=df['sqft_above'], kind='reg')
sns.jointplot(x=df['price'], y=df['grade'], kind='reg')
sns.jointplot(x=df['price'], y=df['bedrooms'], kind='reg')
sns.jointplot(x=df['price'], y=df['bathrooms'], kind='reg');
```

Больше всего влияют на площадь жилых помещений: общая площадь, качество дома, количество санузлов, а меньшее влияние имеют: год постройки и количество спален
